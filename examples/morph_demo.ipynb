# Cross-regime morph demo (DSBM)

Diffusion Schrödinger Bridge Matching between any two of the seven regimes. Inference-time only — does not add to training cost.

Use this notebook to generate the frame sequence used in the YouTube short and the HF Space gallery.

```python
from afm.core.pipeline import AufschreibePipeline, InferenceConfig
pipe = AufschreibePipeline.from_pretrained('hinanohart/aufschreibesysteme-fm')
frames = pipe.morph(
    prompt='a lighthouse at the edge of a stormy sea',
    regime_a='photograph',
    regime_b='gramophone',
    steps=50,
)
print('frames:', len(frames))
```

```python
from PIL import Image
from pathlib import Path
out = Path('morph_out'); out.mkdir(exist_ok=True)
for i, x in enumerate(frames):
    x = x.detach().cpu().float().clamp(0,1)
    if x.ndim == 4: x = x[0]
    arr = (x.permute(1,2,0).numpy()*255).astype('uint8')
    Image.fromarray(arr).save(out / f'frame_{i:04d}.png')
print('wrote', len(frames), 'frames to', out)
```
